```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
F = 15 * (2 ** 10) * (2 ** 10) # kbit
u_s = 30 * (2 ** 10) # kbit/s
d_min = 2 * (2 ** 10) # kbit/s
u_node = np.array([300, 700, 2 * (2 ** 10)]) # kbit/s
N = np.array([10, 100, 1000])
```

```python
MAX_N = 1000
N = np.linspace(0, MAX_N, MAX_N)

```

```python
def distribution_time_ClientServer(N, F, u_s, d_min):
    t1 = N * F / u_s
    t2 = F / d_min

    res = [max(t1[i], t2) for i in range(len(N))]
    
    return res
```

```python
def distribution_time_p2p(N, F, u_s, d_min, u_node):
    t1 = F / u_s
    t2 = F / d_min
    t3 = (N * F) / (u_s + u_node * N)
    
    res = np.array([max(t1, t2, t3[i]) for i in range(len(N))])
    return res
```

```python
y_clientServer = distribution_time_ClientServer(N, F, u_s, d_min)
```

```python
y_p2p_1 = distribution_time_p2p(N, F, u_s, d_min, u_node[0])
y_p2p_2 = distribution_time_p2p(N, F, u_s, d_min, u_node[1])
y_p2p_3 = distribution_time_p2p(N, F, u_s, d_min, u_node[2])
```

```python
plt.plot(N, y_clientServer, label='client-server')
plt.plot(N, y_p2p_1, label=f'p2p u_i={u_node[0]}kb/s')
plt.plot(N, y_p2p_2, label=f'p2p u_i={u_node[1]}kb/s')
plt.plot(N, y_p2p_3, label=f'p2p u_i={u_node[2]}kb/s')


plt.xlabel("N pears")
plt.ylabel("second")
plt.legend(loc="upper left")


plt.show()
```
